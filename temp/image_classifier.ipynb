```python
def classify(img_path, classifier):
    import os
    import numpy as np
    import tensorflow as tf
    import numpy as np
    import os
    from tensorflow.keras.models import load_model
    from tensorflow.keras.preprocessing import image

    class_names = ['FCC', 'IR', 'Optical', 'SAR']  # Adjust this as per your class lab

    if not os.path.exists(img_path, classifier):
        raise FileNotFoundError(f"Image not found: {img_path}")

        # --- Load & preprocess ---
    img_size = (224, 224)
    img = image.load_img(img_path, target_size=img_size)  # Load image and resize
    img_array = image.img_to_array(img).astype("float32")  # Convert image to array
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

        # --- Predict ---
    preds = classifier.predict(img_array)  # Get predictions
    pred_class = class_names[np.argmax(preds[0])]  # Get the predicted class
    confidence = float(np.max(preds[0]))  # Get the confidence of the prediction

    return pred_class

```
